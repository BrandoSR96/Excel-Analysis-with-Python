```python
from google.cloud import bigquery

ruta_conexion = "credenciales/datapath-bigquery.json"

client = bigquery.Client.from_service_account_json(ruta_conexion)
```

```python
query_job = client.query(
    """
    SELECT *
    FROM dap12-datapath-418401.ds_ventas_datapath.ventas
   """)
results = query_job.result()
```

```python
lista_results=[]
for element in results:
    lista_results.append(list(element))
```

```python
lista_results
```

```python
import pandas as pd

df_ventas = pd.DataFrame(lista_results,
                         columns=["fecha_matricula",
                                  "nombreCliente",
                                  "programa",
                                  "nombreVendedor",
                                  "inversion",
                                  "montofinalventa",
                                  "plataforma",
                                  "comunityManager",
                                  "costoUSD",
                                  "costoSoles"
                                  ])
```
