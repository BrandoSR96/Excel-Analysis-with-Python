```python
dfClientes[["PERIODO","CODIGOCLIENTE","EDAD","DISTRITO","FLG_CELULAR"]].groupby("PERIODO").agg({
    "CODIGOCLIENTE":"count",
    "EDAD":"mean",
    "DISTRITO":"nunique",
    "FLG_CELULAR":pd.Series.mode
}).reset_index()
```

```python
dfClientes[["PERIODO","DEPARTAMENTO","CODIGOCLIENTE","EDAD","DISTRITO","FLG_CELULAR"]].groupby(["PERIODO","DEPARTAMENTO"]).agg({
    "CODIGOCLIENTE":"count",
    "EDAD":["mean","median"],
    "DISTRITO":["nunique",pd.Series.mode],
    "FLG_CELULAR":pd.Series.mode
}).reset_index()
```
